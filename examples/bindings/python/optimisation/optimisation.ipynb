```python
from vanta_core_py.optimisers import (
    gradient_descent,
    genetic_algorithm,
    particle_swarm,
    GDOptions,
    GAOptions,
    PSOptions
)
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Rosenbrock objective function
def rosenbrock(x):
    return (1 - x[0])**2 + 100.0 * (x[1] - x[0]**2)**2


# Gradient of Rosenbrock objective function
def rosenbrock_grad(x):
    dx = -2 * (1 - x[0]) - 400 * x[0] * (x[1] - x[0]**2)
    dy = 200 * (x[1] - x[0]**2)
    return np.array([dx, dy])
```

```python
# Initial guess and bounds
x0 = np.array([-1.5, 1.5])
lb = np.array([-3.0, -3.0])
ub = np.array([3.0, 3.0])
```

```python
# Gradient descent options
gd_opts = GDOptions()
gd_opts.learning_rate = 1e-3
gd_opts.max_iters = int(1e5)
gd_opts.tolerance = 1e-8

# Run gradient descent
gd_sol = gradient_descent(
    rosenbrock,
    x0,
    rosenbrock_grad,
    gd_opts
)

# Print results
print("Gradient Descent (with analytical gradient):")
print("x:", gd_sol.x)
print("f:", gd_sol.f_val)
print("iters:", gd_sol.iters)
print("converged:", gd_sol.converged)
print()
```

```python
# Gradient descent options
gd_opts = GDOptions()
gd_opts.learning_rate = 1e-3
gd_opts.max_iters = int(1e5)
gd_opts.tolerance = 1e-8

# Run gradient descent
gd_sol = gradient_descent(
    rosenbrock,
    x0,
    None,
    gd_opts
)

# Print results
print("Gradient Descent (with numerical gradient):")
print("x:", gd_sol.x)
print("f:", gd_sol.f_val)
print("iters:", gd_sol.iters)
print("converged:", gd_sol.converged)
print()
```

```python
# Particle swarm options
ps_opts = PSOptions()
ps_opts.n_particles = 40
ps_opts.max_iters = 500
ps_opts.w = 0.7
ps_opts.c1 = 1.5
ps_opts.c2 = 1.5
ps_opts.tolerance = 1e-8

# Run particle swarm optimization
ps_sol = particle_swarm(
    rosenbrock,
    lb,
    ub,
    ps_opts
)

# Print results
print("Particle Swarm")
print("x:", ps_sol.x)
print("f:", ps_sol.f_val)
print("iters:", ps_sol.iters)
print("converged:", ps_sol.converged)
print()
```

```python
# Genetic algorithm options
ga_opts = GAOptions()
ga_opts.population_size = 80
ga_opts.max_generations = 300
ga_opts.mutation_rate = 0.15
ga_opts.crossover_rate = 0.85
ga_opts.tolerance = 1e-6

# Run genetic algorithm optimisation
ga_sol = genetic_algorithm(
    rosenbrock,
    lb,
    ub,
    ga_opts
)

# Print results
print("Genetic Algorithm")
print("x:", ga_sol.x)
print("f:", ga_sol.f_val)
print("iters:", ga_sol.iters)
print("converged:", ga_sol.converged)
print()
```
